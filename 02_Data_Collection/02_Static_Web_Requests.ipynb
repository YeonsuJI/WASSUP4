```python
html_src = resp.text
```

```python
soup = BeautifulSoup(html_src, 'lxml')
```

```python
## quiz  주요 뉴스의 목록들을 추출해보기
taglist = soup.select('.section_strategy a')
taglist
```

```python
taglist = soup.select('.section_strategy li a')
len(taglist)
```

```python
# 첫번째 뉴스 제목 추출
taglist[0].text
```

```python
taglist[0]['href']
```

```python
url = taglist[0]['href']
PATH = url
```

```python
from urllib.parse import urljoin
urljoin(PATH,url)
```

## quiz : 네이버 증권 주요뉴스의 제목과 url 을 모두 추출하기
제목 변수, url 변수 만들기

```python
news_title =[]
news_url = []

for item in taglist:
    title = item.text
    url = urljoin(PATH, item['href'])
    print(title, url)
    news_title.append(title)
    news_url.append(url)
```
